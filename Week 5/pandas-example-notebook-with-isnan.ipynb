### .ISNAN example
 - np.isnan is applied to the numpy arrays of native dtypes, (https://numpy.org/doc/stable/reference/arrays.dtypes.html) AND will error out when applied to object data types... and the example I was using ['Property City'] happens to be "object" in dtypes. 
 - Always check data first!

```python
import numpy as np
```

```python
df['Property City'].dtypes
```

```python
# Convert 'Property City' to float type
df['Property City'] = pd.to_numeric(df['Property City'], errors= 'coerce')
df['Property City'].dtypes
```

```python
#now np.isnan should populate whether the value(s) are NaN(s)
np.isnan(df['Property City'])
```

```python
#Changing dtypes in pandas is usually done with .astype()

df['Property City'] = df['Property City'].astype(float)
```
