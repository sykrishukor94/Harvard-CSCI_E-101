```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Advertising.csv")
df.head()
```

### One quantitative variable

- ***We can look at the distribution of the quantitative variable to understand its min, max, mean, median, and presence of any outliers*** 


```python
fig, ax = plt.subplots(1,1, figsize=(12,8))

ax =  sns.boxplot(x =df["Sales"])
plt.title("box plot Sales")

```

- ***There seems to be no outliers in the sales.***

```python
df2 = pd.read_csv("cookcounty_home_sales.csv")
df2.head()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,8))

ax =  sns.boxplot(x =df2["Sale Price"])
plt.title("box plot home sale price")
```

***The distribution is right skewed, and we can see a lot of outliers***

```python
a =df2[["Sale Price","Basement"]].groupby("Basement").mean()
a
```

```python
b =df2[["Sale Price","Central Heating"]].groupby("Central Heating").describe()
b
```

```python
c =df2[["Sale Price","Bedrooms"]].groupby("Bedrooms").describe()
c
```
