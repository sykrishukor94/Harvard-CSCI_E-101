```python
import pandas as pd

# pip install tiingo
from tiingo import TiingoClient
```

```python
# prepare api key
config = {}

# create a configuration dictionary to store api key
config['api_key'] = 'xxxxxxxxxxxxxxxxxxxxxxxxxxx'
```

```python
# include a session key to enable reuse of the same HTTP Session for all api calls
config['session'] = True

# Initialize TiingoClient object instance with session and api keys
client = TiingoClient(config)
```

```python
Results=client.get_ticker_price("MSFT",fmt='json',
                                     startDate='2021-07-01',
                                     endDate='2021-07-27',
                                     frequency='daily')
```

```python
Results
```

```python
Stockdf = pd.DataFrame(Results)
```
